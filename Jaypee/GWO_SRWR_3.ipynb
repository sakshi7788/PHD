```python
from sklearn.cluster import KMeans
def K_means_clustering(datapoint,s):
    a=s
    v=np.array(a)
    x=v.reshape(3,1)#change for cluster number,dimensions
    km=KMeans(n_clusters=3,init=x, n_init=1)
    km.fit(datapoint)
    score = silhouette_score(datapoint, km.labels_, metric='euclidean')
    print("silhouette score is",score)
    def ClusterIndicesNumpy(clustNum, labels_array):
        return np.where(labels_array == clustNum)[0]
    cluster_1_records=datapoint[ClusterIndicesNumpy(0,km.labels_)]
    cluster_2_records=datapoint[ClusterIndicesNumpy(1,km.labels_)]
    cluster_3_records=datapoint[ClusterIndicesNumpy(2,km.labels_)]
    return [cluster_1_records,cluster_2_records,cluster_3_records]
```

```python
def inter_cluster(data_point):
    total_sum_data_vector=0
    c=data_point
    s1=0
    #print("shape of data vector of individual cluster is",c.shape)
    m,n=c.shape
    #print(m)
    #print(n)
    f=c.transpose()
    #print(f.shape)
    for x in range(0,n):
        p=f[:][x]
        #print(p.shape)
        data_vector=p.reshape(m,1)
        #print("shape of data point of cluster",data_vector.shape)
        total_sum_data_vector+=data_vector
        #print("total sum of individual data vectors",total_sum_data_vector)
        #print("shape of total sum of individual data vectors",total_sum_data_vector.shape)
        mean_cluster=total_sum_data_vector/n
        #print("mean of each cluster",mean_cluster)
        #print("mean of each cluster shape",mean_cluster.shape)
        mean_overall_cluster=data_point/m*n
        
        diff=(mean_cluster-mean_overall_cluster)
        #print("difference of individual cluster with overall difference",diff.shape)
        s1+=np.dot(diff,diff.transpose())
        #print(s1)
        #print("shape of within cluster distance matrix",s1.shape)
        #print("final inter cluster distance matrix",s1.shape)
    #print("sum of inter cluster distance that is Sb is",s1.trace())
        #print(s1.trace().shape)
    inter=s1.trace()
    inter.reshape(1,1)
    #return[inter]
    print("inter",float(np.asarray(inter)))
    return[float(np.asarray(inter))]
```

```python
import numpy as np
import random

def gwo(fitness, max_iter, n, dim):
    rnd = random.Random(0)
    
    # create n random wolves
    population = [ wolf(fitness, dim,i) for i in range(n)]
    print("printing population",population[0].position)
    
    # On the basis of fitness values of wolves
    # sort the population in asc order
    
    population = sorted(population, key = lambda temp: temp.fitness)
    
    
 
    # best 3 solutions will be called as
    # alpha, beta and gaama
    alpha_wolf, beta_wolf, gamma_wolf = copy.copy(population[: 3])
    print("alpha is",alpha_wolf.position)
    print("beta wolf is",beta_wolf.position)
    print("gamma wolf is",gamma_wolf.position)
    
    
    Iter = 0
    while Iter < max_iter:
         
        
 
        # linearly decreased from 2 to 0
        a = 2*(1 - Iter/max_iter)
 
        # updating each population member with the help of best three members
        for i in range(n):
            
            print("poulation is",population[i].position)
            A1, A2, A3 = a * (2 * rnd.random() - 1), a * (
              2 * rnd.random() - 1), a * (2 * rnd.random() - 1)
            C1, C2, C3 = 2 * rnd.random(), 2*rnd.random(), 2*rnd.random()
            
            
            X1 = [0.0 for i in range(dim)]
            X2 = [0.0 for i in range(dim)]
            X3 = [0.0 for i in range(dim)]
            Xnew = [0.0 for i in range(dim)]
            print("Type of Xnew",type(Xnew))
            
            
            for j in range(dim):
                print("alpha type",type(alpha_wolf.position[j]))
                print("population",population[i].position[j])
                print("alpha",(alpha_wolf.position[j]))
                print("population type",type(population[i].position[j]))
                
                X1[j] = alpha_wolf.position[j] - A1 * abs(
                  C1*alpha_wolf.position[j] - population[i].position[j])
                X2[j] = beta_wolf.position[j] - A2 * abs(
                  C2 *beta_wolf.position[j] - population[i].position[j])
                X3[j] = gamma_wolf.position[j] - A3 * abs(
                  C3*gamma_wolf.position[j] - population[i].position[j])
                Xnew[j]+= X1[j] + X2[j] + X3[j]
                
            for j in range(dim):
                Xnew[j]/=3.0
             

            # fitness calculation of new solution
            fnew = fitness(Xnew)
 
            # greedy selection
            if fnew < population[i].fitness:
                population[i].position = Xnew
                population[i].fitness = fnew
                
        # On the basis of fitness values of wolves
        # sort the population in asc order
        population = sorted(population, key = lambda temp: temp.fitness)
 
        # best 3 solutions will be called as
        # alpha, beta and gaama
        alpha_wolf, beta_wolf, gamma_wolf = copy.copy(population[: 3])
        print("Iteration number is",Iter)
        Iter+= 1
    # end-while
 
    # returning the best solution
    return alpha_wolf.position
           
#----------------------------
 
 
```
