```python
import pandas as pd
```

```python
df = pd.read_csv("BB - slashdot.csv")
df['Name'] = df['Name'].astype(str)
df['Value'] = df['Value'].astype(str)
df
```

```python
df = df.groupby(['Name'])['Value'].apply(list).reset_index()
df
```

```python
df['degree']  = df['Value'].str.len()
```
