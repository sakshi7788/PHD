```python
import numpy as np

from sklearn.cluster import DBSCAN
from sklearn import metrics
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
```

```python
db = DBSCAN(eps=0.3, min_samples=10).fit(df)
```

```python
labels = db.labels_

# Number of clusters in labels, ignoring noise if present.
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
n_noise_ = list(labels).count(-1)

print("Estimated number of clusters: %d" % n_clusters_)
print("Estimated number of noise points: %d" % n_noise_)
print("Homogeneity: %0.3f" % metrics.homogeneity_score(Arr, labels))
print("Completeness: %0.3f" % metrics.completeness_score(Arr, labels))
print("V-measure: %0.3f" % metrics.v_measure_score(Arr, labels))
print("Adjusted Rand Index: %0.3f" % metrics.adjusted_rand_score(Arr, labels))
print(
    "Adjusted Mutual Information: %0.3f"
    % metrics.adjusted_mutual_info_score(Arr, labels)
)
#print("Silhouette Coefficient: %0.3f" % metrics.silhouette_score(df, labels))
```
